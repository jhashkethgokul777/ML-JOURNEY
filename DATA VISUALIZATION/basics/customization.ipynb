```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4,5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, linestyle='--', color='green', marker='o')
plt.title("customized line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```
