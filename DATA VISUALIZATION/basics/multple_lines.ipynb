```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4,5] 

y1 = [1, 4, 9, 16, 25]
y2 = [2, 4, 6, 8, 10]

plt.plot(x, y1, label='squares', color='blue', marker='o')
plt.plot(x, y2, label='Doubles', color='orange', linestyle='--', marker='s')
plt.title("Multiple lines")
plt.grid(True)
plt.legend()
plt.show()
```
