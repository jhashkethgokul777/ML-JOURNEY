```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y, color='blue', marker='o')
plt.title('Saving plot example')
plt.grid(True)

plt.savefig('my_plot.png')
plt.show()
```
