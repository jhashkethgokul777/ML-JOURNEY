```python
import matplotlib.pyplot  as plt
```

```python
categories = ['maths', 'science', 'english', 'history']
values = [80, 90, 70, 85]
colors = ['red', 'blue', 'green', 'orange']

plt.barh(categories, values, color=colors, height=0.6)
plt.title("Subjects VS Marks")
plt.grid(axis='x')
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.show()
```

```python
players = ['Rohit', 'virat', 'gill', 'dhoni']
runs = [80, 110, 95, 120]
colors = ['skyblue', 'purple', 'green', 'orange']

plt.barh(players, runs, color=colors, height=0.5 )
plt.title("Players vs Scores")
plt.xlabel("Scores")
plt.ylabel("players")
plt.grid(axis='x')
plt.show()
```
