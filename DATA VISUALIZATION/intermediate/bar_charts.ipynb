```python
import matplotlib.pyplot as plt
```

```python
categories = ['Apple', 'Banana', 'Orange', 'Mango']
values = [30, 45, 10, 22]

plt.bar(categories, values, color='skyblue', width=0.5)
plt.title("Fruit Sales")
plt.xlabel("Fruits")
plt.ylabel('Quantity')
plt.grid(axis='y')
plt.show()
```
