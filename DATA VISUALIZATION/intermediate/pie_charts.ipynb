```python
import matplotlib.pyplot as plt
```

```python
labels = ['Apple', 'Banana', 'Mango', 'Grapes']
sizes = [40, 25, 20, 15]
colors = ['red', 'yellow', 'orange', 'purple']

plt.pie(sizes, labels=labels, colors=colors, explode=[0.1, 0, 0, 0], autopct='%1.1f%%')
plt.title("Fruit Marker Share")
plt.axis('equal')
plt.show()
```
