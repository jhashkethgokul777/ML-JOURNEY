```python
import matplotlib.pyplot as plt
```

```python
days = [1, 2, 3, 4, 5]
math = [30, 40, 50, 60, 70]
science = [20, 30, 40, 50, 60]
english = [10, 20, 30, 40, 50]

plt.stackplot(days, math, science, english, labels=['math', 'science', 'english'], colors=['blue', 'green', 'red'])

plt.title('Study Hours Over Days')
plt.xlabel('Day')
plt.ylabel('Hours Studied')
plt.legend(loc='upper left')
plt.show()
```
