```python
import matplotlib.pyplot as plt
```

```python
ages = [22, 25, 27, 29, 30, 32, 35, 36, 38, 40, 42, 45, 48, 50, 52]

plt.hist(ages, bins=5, color='skyblue', edgecolor='black', alpha=0.8)
plt.title('Age Distribution')
plt.xlabel('Age Distribution')
plt.ylabel('Number of people')
plt.grid(True)
plt.show()
```

```python
scores = [55, 60, 62, 65, 67, 70, 72, 75, 77, 80, 82, 85, 87, 90, 95]

plt.hist(scores, color='orange', bins=4, edgecolor='black')
plt.title('Score Distribution')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```
