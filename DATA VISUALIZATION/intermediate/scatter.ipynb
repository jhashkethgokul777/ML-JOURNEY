```python
import matplotlib.pyplot as plt
```

```python
x =  [1, 2, 3, 4, 5]
y = [10, 12, 20, 25, 30]

plt.scatter(x, y, color='green', marker='o', s=100)
plt.title('Simple Scatter Plot')
plt.xlabel('x values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()
```
