plotting real datasets with pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#simple sales data

data = {
    'month': ['jan', 'feb', 'mar', 'apr'],
    'Revenue': [1200, 1500, 1800, 2000],
    'profit': [300, 400, 450, 500]
}

df = pd.DataFrame(data)
df
```

plot-1: Line plot from pandas

```python
df.plot(x='month', y=['Revenue', 'profit'], kind='line', marker='o', title='Monthly Revenue vs profit')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.grid(True)
plt.show()
```

plot-2: Bar Chart

```python
df.plot(x='month', y='Revenue', kind='bar', color='skyblue', title='Monthly revenue')
plt.xlabel('month')
plt.ylabel('Revenue')
plt.show()
```

plot-3: pie Chart (profit distribution)

```python
df.plot.pie(y='profit', labels=df['month'], autopct='%1.1f%%', title='Profit Distribution', legend=False )
plt.ylabel("")
plt.show()
```

Task

```python
df = pd.DataFrame({
    'subject': ['math', 'english', 'science', 'history'],
    'marks': [88, 75, 90, 85]
})
```

```python
#bar chart
df.plot(x='subject', y='marks', kind='bar', color='skyblue', title='subject vs marks')
plt.xlabel('subjects')
plt.ylabel('marks')
plt.legend(loc='upper left')
plt.grid(axis='y')
plt.show()
```

```python
#pie chart
df.plot.pie(y='marks', labels=df['subject'], autopct='%1.1f%%', title='Marks Distribution', legend=False)
plt.ylabel("")
plt.show()
```
