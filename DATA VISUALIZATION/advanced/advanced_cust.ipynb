```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color='teal', marker='o')
plt.title('custom plot', fontsize=16, fontweight='bold', color='black')
plt.xlabel('x-axis', fontsize=12)
plt.ylabel('y-axis', fontsize=12)

plt.xticks(fontsize=10, color='red')
plt.yticks(fontsize=10, color='blue')

plt.gca().spines['top'].set_visible(False) #plot boders
plt.gca().spines['right'].set_visible(False) 
plt.gca().spines['left'].set_color('orange')
plt.gca().spines['left'].set_linewidth(2)

plt.annotate("Mid point", xy=(3, 6), xytext=(3.8, 7), arrowprops=dict(facecolor='black', arrowstyle='->')) #fucking amazing
plt.show()
```
