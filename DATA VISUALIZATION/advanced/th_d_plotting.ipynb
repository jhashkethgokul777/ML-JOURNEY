step-1: Import 3D Toolkit

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

1. 3D Line Plot

```python
fig = plt.figure() #creates a new figure window
ax = fig.add_subplot(111, projection='3d') #Adds a subplot to the figure.
    #111 = 1 row, 1 column, 1st plot
    #projection turns into a 3d plot

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9 , 16]
z = [10, 20, 30, 40, 50]

ax.plot(x, y, z, color='blue')
ax.set_title('3D Line Plot')
ax.set_xlabel('x')
ax.set_zlabel('z')
ax.set_ylabel('y')
plt.show()
```

2. Scatter plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.randint(0, 10, 10)
y = np.random.randint(0, 10, 10)
z = np.random.randint(0, 10, 10)

ax.scatter(x, y, z, color='green', s=50)
ax.set_title('3D scatter plot')
plt.show()
```

3. 3D surface plot(Advanced)

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('3D surface plot')

plt.show()
```
