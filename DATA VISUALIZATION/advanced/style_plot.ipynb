```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')

x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.plot(x, y, marker='o') #applies the styles
plt.title('Styled Plot with ggplot')
plt.grid(True)
plt.show()
```
