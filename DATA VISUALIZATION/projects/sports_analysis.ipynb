```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame({
    'match': [f'match {i}' for i in range(1, 11)],
    'runs': [45, 60, 35, 80, 25, 55, 40, 70, 90, 50],
    'balls faced': [30, 45, 28, 50, 22, 35, 30, 48, 55, 33],
    'fours': [4, 6, 3, 8, 2, 5, 4, 7, 9, 4],
    'sixes': [1, 2, 1, 3, 0, 1, 2, 2, 3, 1],
    'strike rate': [150.0, 133.3, 125.0, 160.0, 113.6, 157.1, 133.3, 145.8, 163.6, 151.5]
})
```

```python
#Line plots - Runs over matches
x = df['match']
y = df['runs']
plt.plot(x, y, marker='o')
plt.title(' Runs over matches')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
#2. bar plot - strike rate per match
x = df['match']
y = df['strike rate']
plt.bar(x, y, color=['orange', 'yellow', 'teal', 'purple', 'red', 'skyblue', 'green', 'blue', 'violet', 'black'], width=0.6)
plt.xlabel('matches', fontsize=15)
plt.ylabel('strike rate', fontsize=15)
plt.xticks(fontsize=9, rotation=45)
plt.show()
```

```python
#3. stack plots
df.plot.bar(x='match', y=['fours', 'sixes'], stacked=True)

plt.title('Fours and sixes per match')
plt.xlabel('match')
plt.ylabel("count")
plt.tight_layout()
plt.legend(title='Shot Type')
plt.show()
```
