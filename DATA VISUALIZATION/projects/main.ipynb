```python
import pandas as pd
import matplotlib.pyplot as plt
```

create a data visualization report using different types of plots

```python


data  = {
    'student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'math': [88, 92, 95, 70, 85],
    'science': [90, 85, 82, 88, 76],
    'english': [75, 78, 80, 72, 90],
    'Total Study Hours': [30, 40, 38, 20, 35]
}

df = pd.DataFrame(data)
```

1. Bar chart - total marks by student

```python
df['total marks'] = df[['math', 'science', 'english']].sum(axis=1)
plt.bar(df['student'], df['total marks'], color='skyblue')
plt.xlabel('students', color='teal')
plt.ylabel('total marks', color='teal')
plt.title('Total marks by students', fontweight='bold')
plt.grid(axis='y')
plt.show()
```

2. line plot

```python
x = df['student']
y1 = df['math']
y2 = df['science']
y3 = df['english']
plt.plot(x, y1, label = 'math', marker='o')
plt.plot(x, y2, label='science', marker='s')
plt.plot(x, y3, label='english', marker='^')
plt.title('Subject-Wise performance', fontweight='bold')
plt.xlabel('students')
plt.ylabel('marks')
plt.grid(True)
plt.legend()
plt.show()
```

3. pie Chart-study hours distribution

```python
labels = df['student']
values = df['Total Study Hours']

plt.pie(values, labels=labels, autopct='%1.1f%%', explode=[0, 0.1, 0, 0, 0])
plt.title('Study hour distribution')
plt.show()
```

4. 3D scatter plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = df['Total Study Hours']
y = df['math']
z = df['science']

ax.scatter(x, y, z, s=100)
ax.set_title('Study hours vs marks')
ax.set_xlabel('Total study hours')
ax.set_ylabel('Math')
ax.set_zlabel('science')
plt.show()
```
